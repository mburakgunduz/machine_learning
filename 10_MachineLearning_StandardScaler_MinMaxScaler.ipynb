```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
df=pd.read_csv('plane.csv')
df.head(3)
```

```python
df.info()
```

```python
df=df[["Rcmnd cruise Knots","Stall Knots dirty","Fuel gal/lbs","Eng out rate of climb","Price"]]
```

```python
df.head(3)
df=df.dropna()
```

```python
y=df['Price']
x=df.drop("Price",axis=1)
```

```python
# Y için normalizasyon yapılmaz

# outlier etkisini azaltır

# model performansı artar
```

```python
ss=StandardScaler()
x2=ss.fit_transform(x)
```

```python
np.mean(x2)
```

```python
x2=pd.DataFrame(x2)
x2.head(3)
```

```python
x2[1].mean()
```

```python
x2[1].std()
```

```python
mm=MinMaxScaler()
```

```python
x3=mm.fit_transform(x)
```

```python
x3=pd.DataFrame(x3)
```

```python
x3.head(3)
```
